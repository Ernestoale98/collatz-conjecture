```python
from collatz import get_collatz_numbers

numbers = get_collatz_numbers(27)
indexs = [index for index,number in enumerate(numbers)]
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
fig.suptitle('Collatz Numbers')

ax = plt.axes()
ax.set(xlabel='Iteration', ylabel='Value')


plt.plot(indexs, numbers, marker='o', linestyle='dashed',
         linewidth=2, markersize=8)
```
